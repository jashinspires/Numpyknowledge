# Numpy Array vs Python List (**Time comparision** , not **Time complexity**)

```python
# Speed of list vs Speed of n array , Referential Array , Dynamic in size
a = [i for i in range(10000000)]
b = [i for i in range(10000000,20000000)]

c = []
import time
start = time.time()
for i in range(len(a)):
  c.append(a[i] + b[i])
print(f"{time.time() - start}")
```

```python
import numpy as np # fast because it directly stores values in memory not as a reference , and also it is static size
import time
a1 = np.arange(10000000)
b1 = np.arange(10000000, 20000000)

s1 = time.time()
c1 = a1 + b1
print(f"NumPy time: {time.time() - s1}")
```

# **BroadCasting**

- The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations
- The smaller array is "broadcast" across the larger array so that they have compatible shapes


```python
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)
print(a)
print(b)
print(a+b)
```

# BroadCasting Rules

 #### 1. Make the two arrays have the same number of dimensions.
 - if the numbers of dimensions of the two arrays are different , add new dimensions with size 1 to the head of the array with the smaller dimensions.
#### 2. Make each dimension of the two arrays the same size.
- If the sizes of each dimension of the two arrays do not match, dimensions with size 1 are stretched to the size of the other.
- if there is a dimension whose size is not 1 in either of the two arrays , it cannot be broadcasted , and an error is raised.
